```python
# For more complex data, we can use a DataFrame. In a DataFrame, each value is a Series of numbers.
# We can use a two-dimensional DataFrame to handle many  exam scores for each key/student.
# Create a new file to practice using DataFrames. 
```

```python
# Follow the instructions in 7.14.2 to import pandas as pd and create a pandas DataFrame from a dictionary (a set of key-value pairs) 
# holding multiple exam scores for each key/student. In this example, we have 5 students, each with 3 exam scores, but the scores aren't labeled.
import pandas as pd
grades_dict = {'Wally': [87, 96, 70], 'Eva': [100, 87, 90], 'Sam': [94, 77, 90], 'Katie': [100, 81, 82], 'Bob': [83, 65, 85]}
grades = pd.DataFrame(grades_dict)
print(grades)

# Follow the instructions to apply a custom index (instead of 0,1,2), let's call them 'Test1', Test2', and 'Test3'. 
# See 'Customizing a DataFrame's Indices with the index Attribute subsection. 
grades.index = ['Test1', 'Test2', 'Test3']
print(grades)

# As we did in Part 1, we can request grades by key using the text attribute (square brackets with single quoted text strings). Find Eva's grades with grades['Eva'] 
print(grades['Eva'])

# If no spaces in the key, we can use the simpler dot attribute approach. Find Sam's grades with grades.Sam as shown in 'Accessing a DataFrame's columns' subsection.
print(grades.Sam)

# Like spreadsheets, we can access specific rows or columns. Execute the examples using loc['Test1'] to get scores for the first exam, 
# or iLoc[0] to get scores for the first exam and iLoc[1] to get scores for the second exam. Which do you prefer? 
print(grades.loc['Test1'])
print(grades.iloc[0])
print(grades.iloc[1])            

# We can use this same notation to get a single cell in the DataFrame (much like getting a single cell in a spreadsheet). 

# Use grades.at['Test2', 'Eva'} to find her score on the second exam.
print(grades.at['Test2', 'Eva'])

# Use grades.iat[2,0] to get the score on the third exam for the first student (Wally). 
print(grades.iat[2, 0])

# Use grades.describe() to get descriptive statistics for our gradebook.  See why learning Python and DataFrames can be so powerful?  :)
print(grades.describe())

# Get the average score for each student key by calling grades.mean()
print(grades.mean())

# Sort the gradebook rows in reverse order so the most recent exam row appears at the top with grades.sort_index(ascending=False)
print(grades.sort_index(ascending=False))

# Sort the gradebook columns so the names appear in order using grades.sort_index(axis=1).
print(grades.sort_index(axis=1))

# We can sort our data pretty much however we like. 
# Take a screenshot showing your work as you finish this part of the project.
print()
print('Amanda Hanway')
```
